### (a)

```python
data1112 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DEMO_G.XPT')
data1314 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT')
data1516 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.XPT')
data1718 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DEMO_J.XPT')

data1112 = data1112[['SEQN','RIAGENDR','RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL', 
                    'RIDSTATR', 'SDMVPSU', 'SDMVSTRA', 'WTMEC2YR', 'WTINT2YR']]
data1112["years"] = "2011-2012"
data1314 = data1314[['SEQN', 'RIAGENDR','RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL', 
                    'RIDSTATR', 'SDMVPSU', 'SDMVSTRA', 'WTMEC2YR', 'WTINT2YR']]
data1314["years"] = "2013-2014"
data1516 = data1516[['SEQN', 'RIAGENDR','RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL', 
                    'RIDSTATR', 'SDMVPSU', 'SDMVSTRA', 'WTMEC2YR', 'WTINT2YR']]
data1516["years"] = "2015-2016"
data1718 = data1718[['SEQN', 'RIAGENDR','RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL', 
                    'RIDSTATR', 'SDMVPSU', 'SDMVSTRA', 'WTMEC2YR', 'WTINT2YR']]
data1718["years"] = "2017-2018"
df = pd.concat([data1112, data1314, data1516, data1718])

df = df.rename(columns={'SEQN': 'id',
    'RIDAGEYR': 'age',
    'RIAGENDR': 'gender',
    'RIDRETH3': 'race',
    'DMDEDUC2': 'education',
    'DMDMARTL': 'marital_status',
    'RIDSTATR': 'exam_status',
    'SDMVPSU': 'psu',
    'SDMVSTRA': 'strata',
    'WTMEC2YR': 'exam_wt',
    'WTINT2YR': 'interview_wt'})


df['id'] = df.id.astype('int')
df['age'] = df.age.astype('int')
df['psu'] = df.psu.astype('int')
df['strata'] = df.strata.astype('int')
df['years'] = pd.Categorical(df['years'])
df['gender'] = pd.Categorical(df['gender'].replace({1: 'Male', 2: 'Female'}))
df['race'] = pd.Categorical(df['race'].replace({1: 'Mexican American',
             2: 'Other Hispanic',
             3: 'Non-Hispanic White',
             4: 'Non-Hispanic Black',
             6: 'Non-Hispanic Asian',
             7: 'Other/Multiracial'}))
df['education'] = pd.Categorical(df['education'].replace({1: 'Less than 9th grade',
                  2: '9-11th grade (Includes 12th grade with no diploma)',
                  3: 'High school graduate/GED or equivalent',
                  4: 'Some college or AA degree',
                  5: 'College graduate or above',
                  7: 'Refused',
                  9: "Don't know"}))
df['marital_status'] = pd.Categorical(df['marital_status'].replace({1: 'Married',
                       2: 'Widowed',
                       3: 'Divorced',
                       4: 'Separated',
                       5: 'Never married',
                       6: 'Living with partner',
                       77: 'Refused',
                       99: "Don't know"}))
df['exam_status'] = pd.Categorical(df['exam_status'].replace({1: 'Interviewed only',
                    2: 'Both interviewed and MEC examined'}))


print(df.dtypes)
df.to_pickle("./3a.pkl")
df
```

### (b)

```python
oral1112 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/OHXDEN_G.XPT')
oral1314 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/OHXDEN_H.XPT')
oral1516 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/OHXDEN_I.XPT')
oral1718 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/OHXDEN_J.XPT')
```

```python
oral = pd.concat([oral1112, oral1314, oral1516, oral1718])
oral1 = oral.filter(regex='TC')                           
oral1 = oral1[oral1.columns.drop(list(oral1.filter(regex='RTC')))]
oral2 = oral[['SEQN', 'OHDDESTS']]
result = pd.concat([oral2, oral1], axis=1)
result = result.rename(columns={"SEQN": "id", "OHDDESTS": "ohx_status"})
result['id'] = result.id.astype('int')
result['ohx_status'] = pd.Categorical(result['ohx_status'].replace({1: 'Complete', 
                                                                    2: 'Partial', 
                                                                    3: 'Not Done'}))

for i in range(1,33):
    if i < 10:
        i = str(i)
        i = '0' + i
    else:
        i = str(i)
    a = 'OHX{}TC'.format(i)
    b = 'teethcount{}'.format(i)
    result = result.rename(columns={a: b})

for j in range(2,32):
    if j < 10:
        j = str(j)
        j = '0' + j
    else:
        j = str(j)
    a = 'OHX{}CTC'.format(j)
    b = 'coronal_cavities{}'.format(j)
    result = result.rename(columns={a: b})
    
result.to_pickle("./3b.pkl")
result
```
